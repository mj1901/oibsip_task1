```python
import pandas as pd

# Load the retail sales dataset
menu_data = pd.read_csv('menu.csv')
```

```python
# Display the first few rows of the dataset
print(menu_data.head())
```

```python
# Check for missing values
print(menu_data.isnull().sum())
```

```python
# Handle missing values if any
# For example, you can fill missing values with mean or median for numerical columns
menu_data.fillna(menu_data.mean(), inplace=True)
```

```python
# Check for duplicates and remove if any
menu_data.drop_duplicates(inplace=True)
```

```python
# Calculate basic statistics
mean_values = menu_data.mean()
median_values = menu_data.median()
mode_values = menu_data.mode().iloc[0]
std_dev_values = menu_data.std()

# Display the calculated statistics
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\nStandard Deviation:\n", std_dev_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the retail sales dataset
menu_data = pd.read_csv('menu.csv')

# Create a numeric index for x-axis (assuming a continuous sequence)
numeric_index = range(len(menu_data))

# Plotting sales trends over data points with advanced effects
plt.figure(figsize=(14, 8))

# Plotting the total calories sold
plt.plot(numeric_index, menu_data['Calories'], label='Total Calories', color='blue', marker='o', linestyle='-')

# Adding a second y-axis for additional information (example: Protein)
ax2 = plt.gca().twinx()
# Plotting the average protein
ax2.plot(numeric_index, [menu_data['Protein'].mean()] * len(numeric_index), label='Average Protein', color='green', marker='s', linestyle='--')

# Adding grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adding title and labels
plt.title('Total Calories and Average Protein Trends Over Data Points', fontsize=16)
plt.xlabel('Data Points', fontsize=14)
plt.ylabel('Total Calories Sold', fontsize=14)
ax2.set_ylabel('Average Protein', fontsize=14)

# Adding legend
lines, labels = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2, loc='upper left')

# Show plot
plt.show()
```

```python
# Analyze customer demographics and purchasing behavior
customer_demographics = menu_data[['Category', 'Calories', 'Protein', 'Carbohydrates', 'Sugars', 'Total Fat']]
print(customer_demographics.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming menu_data is your DataFrame

# Ensure the index is in datetime format
menu_data.index = pd.to_datetime(menu_data.index)

# Visualization using bar charts with Seaborn
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=menu_data, palette='viridis')
plt.title('Distribution of Items by Category')
plt.xlabel('Category')
plt.ylabel('Number of Items')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()

# Resample the data and fill NaN values with 0
resampled_data = menu_data.resample('M').mean().fillna(0)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming menu_data is your DataFrame

# Visualization using line plot with Seaborn
plt.figure(figsize=(12, 6))

# Create a numeric index for x-axis (assuming a continuous sequence)
numeric_index = range(len(menu_data))

# Plotting the data
sns.lineplot(x=numeric_index, y=menu_data['Calories'], label='Average Calories', color='skyblue')

# Enhancements: title, labels, and legend
plt.title('Average Calories Over Data Points')
plt.xlabel('Data Points')
plt.ylabel('Calories')
plt.legend()

# Show the plot
plt.show()
```

```python


# Visualization using heatmap with Seaborn
correlation_matrix = menu_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, square=True)
plt.title('Correlation Matrix')
plt.show()
```
